```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-1000, 1000, 1)
y = x**2
plt.plot(x, y)
plt.ticklabel_format(style = 'plain')
plt.show()
```

```python
y1 = x**2
y2 = x**3
y3 = -x**2
y4 = -x**3
plt.figure(figsize = (10,10))
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.subplot(2, 2, 3)
plt.plot(x, y3)
#plt.subplot(2, 2, 4)
#plt.plot(x, y4)
plt.show()
```

```python
figura = plt.figure(figsize = (10, 7))

figura.add_subplot(2, 2, 1)
plt.plot(x, y1)
figura.add_subplot(1, 2, 2)
plt.plot(x, y2)
figura.add_subplot(2, 2, 3)
plt.plot(x, y3)
plt.show()
```

```python
x = np.arange(-1000, 1000, 1)
y1 = x**2
y2 = x**3
y3 = -x**2
y4 = -x**3
y = [y1, y2, y3, y4]

figura = plt.figure(figsize = (10, 7))

for i in range(4):
    figura.add_subplot(2,2, i+1)
    plt.plot(x, y[i])
plt.show()
```
